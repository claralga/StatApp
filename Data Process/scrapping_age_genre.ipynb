On scrape l'âge

```python
import urllib
import bs4
import pandas as pd
from urllib.request import Request, urlopen
from urllib import request, parse
import ssl
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt

liste_age = []

for k in df_age.index:
    context = ssl._create_unverified_context()
    url = 'https://fr.wikipedia.org/'
    path = parse.quote('wiki/' + df_age.loc[k]['wikilink'])
    req = Request(url+path, headers = {'User-Agent': 'Chrome/5.0'})
    webpage = urlopen(req).read() 
    page = bs4.BeautifulSoup(webpage, "lxml")
    age = page.find('span',{'class':'noprint'}).text.strip()[1:3]
    liste_age.append(age)
```

```python
len(liste_age)
```

```python
df_age['age'] = liste_age
```

```python
df_age
```

```python
df_age.to_csv('age.csv',index=True)
```

On scrape le genre

```python
#https://data.assemblee-nationale.fr/acteurs/deputes-en-exercice
```

```python
df_age = pd.read_csv('age.csv')
votants = df_age['idVotant'].unique()
```

```python
#Erreur
votants = list(votants)
votants.remove('PA331567')
votants.remove('PA719806')
votants.remove('PA717151')
votants.remove('PA998')
```

```python
df_acteurs = pd.DataFrame(columns=['idVotant', 'Genre'])
for votant in votants : 
    df_acteur = pd.read_csv('acteur/'+votant+'.csv',sep=";")
    df_acteur = df_acteur[["uid[1]", "etatCivil[1]/ident[1]/civ[1]"]]
    df_acteur = df_acteur.rename(columns={"uid[1]": "idVotant", "etatCivil[1]/ident[1]/civ[1]": "Genre"})
    df_acteurs = df_acteurs.append(df_acteur, ignore_index = True)
```

```python
df_acteurs
```

```python
#Ajout des 4 données manquantes : 
df_age[df_age['idVotant'].isin(['PA331567'])] #Laure
df_acteurs = df_acteurs.append({'idVotant' : 'PA331567', 'Genre' : 'Mme'}, ignore_index = True)

df_age[df_age['idVotant'].isin(['PA719806'])] #Sophie
df_acteurs = df_acteurs.append({'idVotant' : 'PA719806', 'Genre' : 'Mme'}, ignore_index = True)

df_age[df_age['idVotant'].isin(['PA717151'])] #Marielle
df_acteurs = df_acteurs.append({'idVotant' : 'PA717151', 'Genre' : 'Mme'}, ignore_index = True)

df_age[df_age['idVotant'].isin(['PA998'])] #Olivier
df_acteurs = df_acteurs.append({'idVotant' : 'PA998', 'Genre' : 'M.'}, ignore_index = True)
```

```python
df_acteurs.to_csv('genre.csv',index=True)
```

```python
df_acteurs = df_acteurs.merge(df_age, on = 'idVotant')
```
