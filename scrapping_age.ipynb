```python
import urllib
import bs4
import pandas as pd
from urllib.request import Request, urlopen
from urllib import request, parse
import ssl
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt

liste_age = []

for k in df_age.index:
    context = ssl._create_unverified_context()
    url = 'https://fr.wikipedia.org/'
    path = parse.quote('wiki/' + df_age.loc[k]['wikilink'])
    req = Request(url+path, headers = {'User-Agent': 'Chrome/5.0'})
    webpage = urlopen(req).read() 
    page = bs4.BeautifulSoup(webpage, "lxml")
    age = page.find('span',{'class':'noprint'}).text.strip()[1:3]
    liste_age.append(age)
```
