On crée un individu fictif dans chaque parti, qui vote comme la majorité de son parti à chaque scrutin. Le but est de projeter ensuite ces individus fictifs sur l'ACP. Cela nous permettrait d'observer comment ces députés fictifs évoluent les uns par rapport aux autres, mais aussi comment les député.e.s du parti évoluent par rapport à ce député fictif. 

```python
import pandas as pd
```

```python
df_votes = pd.read_csv('database_votes2.csv')
```

```python
copy = df_votes.pivot(index = 'idVotant', columns = 'idScrutin', values = 'vote')
copy = copy.fillna('Non-votant')
```

```python
groupes = df_votes['Groupe politique (complet)'].unique()
```

```python
scrutins_2017 = df_votes[df_votes['date_scrutin'].str.startswith('2017')]['idScrutin'].unique()
scrutins_2018 = df_votes[df_votes['date_scrutin'].str.startswith('2018')]['idScrutin'].unique()
scrutins_2019 = df_votes[df_votes['date_scrutin'].str.startswith('2019')]['idScrutin'].unique()
scrutins_2020 = df_votes[df_votes['date_scrutin'].str.startswith('2020')]['idScrutin'].unique()
```

```python
dep_fictifs_2017 = pd.DataFrame()

for groupe in groupes : 
    votants = df_votes[df_votes['Groupe politique (complet)'].isin([groupe])]['idVotant'].unique()
    df_used = copy.loc[votants]
    df_used = df_used[scrutins_2017]
    vote = list(df_used.value_counts().idxmax())
    dep_fictifs_2017[groupe] = vote
dep_fictifs_2017 = dep_fictifs_2017.T
dep_fictifs_2017
```
